```python
# focus on file
# '12k Drive End Bearing Fault Data'
# '12k Fan End Bearing Fault Data'
# '48k Drive End Bearing Fault Data'

import os, sys
import random

import scipy.io as io
import seaborn
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tqdm import tqdm
# from function_lab import *
from pyts.image import GramianAngularField
from pyts.image import MarkovTransitionField


for setting in ['12k Drive End Bearing Fault Data', '12k Fan End Bearing Fault Data', '48k Drive End Bearing Fault Data']:

    print(setting)

    rootpath1='/content/drive/MyDrive/Colab Notebooks/Bearing Fault Diagnosis/Experiment 1/GAF-MTF DA/' + setting

    # choose DA or without DA here
    FLAG = 'with DA'
    # FLAG = 'without DA'

    if FLAG=='with DA':
        rootpath1 = '/content/drive/MyDrive/Colab Notebooks/Bearing Fault Diagnosis/Experiment 1/with DA/' + setting
        rootpath2='/content/drive/MyDrive/Colab Notebooks/Bearing Fault Diagnosis/Experiment 1/GAF-MTF DA/'+ setting
    elif FLAG=='without DA':
        rootpath1 = '/content/drive/MyDrive/Colab Notebooks/Bearing Fault Diagnosis/Experiment 1/without DA/' + setting
        rootpath2='/content/drive/MyDrive/Colab Notebooks/Bearing Fault Diagnosis/Experiment 1/GAF-MTF without DA/'+ setting

    datafiles=os.listdir(rootpath1)

    for file in tqdm(datafiles):

        data=io.loadmat(os.path.join(rootpath1,file))['data'].reshape(1,-1)

        # GAF
        # print(file)
        gasf = GramianAngularField(image_size=256, method='difference')
        GAFdata = gasf.fit_transform(data)

        # MTF
        # print(file)
        mtf = MarkovTransitionField(image_size=256)
        MTFdata = mtf.fit_transform(data)

        io.savemat(os.path.join(rootpath2,file),{'GAF':GAFdata, 'MTF':MTFdata})



```
